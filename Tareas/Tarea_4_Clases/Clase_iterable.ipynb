```python
import pandas as pd
iris = pd.read_csv("iris.csv")

class SimpleIterable:
 def __init__(self, data):
    self.data = data
    self.index = 0
        
 def __iter__(self):
    return self
    
 def __next__(self):
    if self.index < len(self.data):
        item = self.data[self.index]
        self.index += 1
        return item
    else:
        raise StopIteration
```

```python
# Crear una instancia de la clase SimpleIterable
iterable = SimpleIterable([1, 2, 3, 4, 5])

# Iterar sobre los elementos del iterable
for item in iterable:
    print(item)
```

```python
prueba1 = SimpleIterable(iris["sepal.length"])

for item in prueba1:
   print(item)
```

```python
class IrisDataIterable:
    def __init__(self, file_path):
        self.df = pd.read_csv(file_path)
        self.index = 0
        
    def __iter__(self):
        return self
    
    def __next__(self):
        if self.index < len(self.df):
            next_data = self.df.iloc[self.index]
            self.index += 2
            return next_data
        else:
            raise StopIteration
```

```python
# Crear una instancia de la clase IrisDataIterable
iris_iterable = IrisDataIterable("iris.csv")

# Iterar sobre los datos del archivo "iris.csv"
for data_row in iris_iterable:
    print(data_row)
```
