  <sub><H2>2: Creando la Clase Iterable. </H2></sub>
     
Ahora bien la siguiente clase toma un dataframe de tipo CSV completo, lo almacena dentro de la propia clase y si se le pide, lo itera a los 2 siguientes Datos:

 </body>    
</html>    

```python
#La primera Clase que se nos pide:
class CLASE1:

#En este caso los objetos de la clase solo tendran como atributo
#el DataFrame en formato CSV y lo leera con pandas."        
#Integramos el read_csv de pandas a la clase por medio del _init_:
 def __init__(self, DataFrameCSV):
        self.df = pd.read_csv(DataFrameCSV)
        self.index = 0

#Utilizamos la forma mas simple del método _iter_ para recorrer los elementos
#del DataFrame, evidentemente necesitaremos definir el consecutivo de la iteración:                
 def __iter__(self):
        return self

#_next_ define dicho consecutivo, en este caso, para Dataframes (he ahi la razón del iloc).    
#Modifiqué el "siguiente_dato" para organizar la salida en la iteración:    
 def __next__(self):
    if self.index < len(self.df):
        siguiente_dato = self.df.iloc[self.index:self.index+2]
        self.index += 2
        return siguiente_dato
    else:
        raise StopIteration("error")
```

<html>
<body>
    Probemos nuestra nueva Clase Iteradora con los datos de Iris.csv:

```python
# Ahora solo debemos instanciar el "iris.csv"
iris_iterable = CLASE1("iris.csv")
```

<html>
<body>
    Podremos probar como itera nuestra clase tanto iterando todos los datos dentro del DataFrame o bien únicamente solo los datos correspondientes:
</body>
</html>

```python
# Iterar en únicamente un solo valor, se ira actualizando de 2 en 2 , se realiza de la siguiente forma:
iteracion_1 = next(iris_iterable)
print(iteracion_1)
```

```python
# Iterar y obtener los siguientes dos conjuntos de datos en cada iteración:  
try:
 while True:
        iteracion2 = next(iris_iterable)
        print(iteracion2)
except StopIteration as e:
    if str(e) == "error":
        print("Error: Los datos se han agotado.")       
```
